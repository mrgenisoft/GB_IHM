### Задание 1

Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20).

1.1 Найдите их сумму.

Поэлементное сложение: (20,20,20) + (0,0,-20) = (20,20,0)

1.2 Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.

```python
def vec_len(v):
    assert type(v) == np.ndarray, "Only Numpy array allowed"
    return np.sqrt(np.sum(v * v))
```

### Задание 2

Почему прямые не кажутся перпендикулярными?

Масштаб осей должен быть одинаковым для корректного отображения углов между прямыми.

Также должны быть равны длина и ширина поля отображения графика.

```python
# неверно
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = x / 3 + 1
plt.plot(x, y, label='y = 3 * x + 1')
plt.plot(x, y2, label = 'y = x / 3 + 1')
plt.legend()
plt.show()
```

```python
# верно
fig, ax = plt.subplots()
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = - x / 3 + 1
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_aspect(aspect='equal')
ax.plot(x, y, label='y = 3 * x + 1')
ax.plot(x, y2, label = 'y = - x / 3 + 1')
ax.legend()
plt.show()
```

### Задание 3

Напишите код на Python, реализующий построение графиков: окружности, эллипса, гиперболы.

```python
# окружность
x0 = 5
y0 = 10
R = 20

x = np.linspace(x0-R, x0+R)
y1 = np.sqrt(R**2 - (x - x0)**2) + y0
y2 = - np.sqrt(R**2 - (x - x0)**2) + y0

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
ax.set_aspect(aspect='equal')
plt.show()
```

```python
# эллипс
x0 = 5
y0 = 10
a = 20
b = 5

t = np.linspace(0, 2*np.pi)
x = a * np.cos(t) + x0
y = b * np.sin(t) + y0

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_aspect(aspect='equal')
plt.show()
```

```python
# гипербола
x0 = 50
y0 = 20
a = 10
b = 5

t = np.linspace(-3, 3)
x1 = a * np.cosh(t) + x0
x2 = - a * np.cosh(t) + x0
y = b * np.sinh(t) + y0

fig, ax = plt.subplots()
ax.plot(x1, y)
ax.plot(x2, y)
ax.set_aspect(aspect='equal')
plt.show()
```

### Задание 5

5.1 Нарисуйте трехмерный график двух параллельных плоскостей.

```python
A = 2
B = -3
C = 4
D = 20

x = np.linspace(-5, 5)
y = np.linspace(-5, 5)
x, y = np.meshgrid(x, y)
z1 = (- A * x - B * y - D) / C
z2 = (- A * x - B * y) / C

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(x, y, z1, color='red')
ax.plot_wireframe(x, y, z2, color='blue')
plt.show()
```

5.2 Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
# однополостный гиперболоид
x0 = 0
y0 = 0
z0 = 0
a = 1
b = 1
c = 1

v = np.linspace(0, 2*np.pi)
u = np.linspace(-2, 2)
u, v = np.meshgrid(u, v)
x = x0 + a * np.cosh(u) * np.cos(v)
y = y0 + b * np.sinh(u) * np.sin(v)
z = z0 + c * np.sinh(u)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(x, y, z)
plt.show()
```

```python
# гиперболический параболоид
x0 = 0
y0 = 0
a = 1
b = 1

v = np.linspace(-2, 2)
u = np.linspace(-2, 2)
u, v = np.meshgrid(u, v)
x = x0 + a * (u + v)
y = y0 + b * (v - u)
z = 4 * u * v

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(x, y, z)
plt.show()
```
