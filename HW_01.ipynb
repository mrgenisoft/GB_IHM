### Задание 4

Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
def fun_cos(k, x):
    return np.cos(k * x)
```

```python
x = np.linspace(0, 5)
```

```python
y = fun_cos(k=1, x=x)
plt.plot(x, y, label='k=1')
y = fun_cos(k=3, x=x)
plt.plot(x, y, label='k=3')
plt.legend()
plt.show()
```
